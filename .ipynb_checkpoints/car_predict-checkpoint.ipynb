# Splitting the data for training and test

```python
X = car_data.drop("selling_price", axis = 1)
Y = car_data["selling_price"]
print(X)
print(Y)
```

```python
# Split the data into training set and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
# print(X_train)
# print(Y_train)
# # Initialize the StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
# X_train_scaled = pd.DataFrame(X_train_scaled)
# X_test_scaled = pd.DataFrame(X_test_scaled)
```

# Train the Model

```python
model = LinearRegression()
model.fit(X_train_scaled, Y_train)
```

```python
# Make predictions on the scaled test data
Y_test_pred = model.predict(X_test_scaled)
```

## Evaluate the model

```python
# Calculate the Mean Squared Error (MSE) and R-squared
mse = mean_squared_error(Y_test, Y_test_pred)
r2 = r2_score(Y_test, Y_test_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
# Create a scatter plot to compare actual vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(Y_test, Y_test_pred, color='blue', alpha=0.6)

# Adding labels and title
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Car Prices (Y_test)')
plt.ylabel('Predicted Car Prices (Y_test_prediction)')

# Show a line where actual = predicted for reference
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], color='red', linestyle='--')

# Show the plot
plt.show()
```
